```python
#coffee machine
from main import MENU,resources

def report():
    print(f"Water: {resources['water']}ml")
    print(f"Milk: {resources['milk']}ml")
    print(f"Coffee: {resources['coffee']}g")
    print(f"Money: ₹{resources['money']}")
    
from main import MENU,resources

user=input("what do you want? espresso,latte,cappuccino")
if user=="report":
    report()
    
else:
    drink=MENU[user]
    ingredients=drink["ingredients"]


    can_make=True
    for item in ingredients:
        if ingredients[item]>resources[item]:
            print(f"cannot make.insufficient {item}")
            can_make=False

    if can_make:
        payment=int(input(f"Please Pay for the drink.Your bill is {MENU[user]['cost']}:"))
    
        if payment>=MENU[user]["cost"]:
            if payment>MENU[user]["cost"]:
                print(f"Here is your exchange {payment-MENU[user]["cost"]}")
            
            for item in ingredients:
                resources[item]-=ingredients[item]
        
            resources["money"] += MENU[user]["cost"]
            print("Here is your coffee.Enjoy")
        else:
            print("Not enough money")

```

```python
report()
```
